What was the most frequent movie duration in the 1990s? Save an approximate answer as an integer called **duration** (use 1990 as the decade's start year).

```python
# importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
netflix_df.head()
```

```python
# Looking into the csv file in depth
netflix_df.info()

#Rows and columns
netflix_df.shape
```

```python
# checking for missing values
netflix_df.isnull().sum()
```

```python
# Frequent movies from 1990 tp 1999
'''
Steps:
1. Select the release_year column.
2. Do comparisions on the column.
3. Use the result to select movies from 1990 to 1999.
'''

netflix_df["release_year"]
```

```python
netflix_subset = netflix_df[netflix_df["type"] == "Movie"]
```

```python
movies_1990 = netflix_subset[(netflix_subset["release_year"] >= 1990) & (netflix_subset["release_year"] <2000)]
```

```python
# plotting duration using matplotlib

plt.hist(movies_1990["duration"])
plt.xlabel("Duration")
plt.ylabel("No of movies")
plt.title("Number of movies from 1990 to 1999")
plt.show()

duration = 100
```

```python
# Filtering movies from 1990 where the length of the action movies < 90 minutes.
action_movies= movies_1990[movies_1990["genre"] == "Action"]

short_movie_count = 0

for lable, row in action_movies.iterrows():
    if row["duration"] < 90:
        short_movie_count = short_movie_count + 1
    else:
        short_movie_count = short_movie_count
        
print(short_movie_count)
```
